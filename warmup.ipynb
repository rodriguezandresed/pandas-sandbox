```python
bios['heigh_category']=bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
#the apply method allows us to apply a function to each row or column of the dataframe
#in this case we are using a lambda function to categorize the height into Short, Average or Tall
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'
    
    bios['Category']=bios.apply(categorize_athlete, axis=1)
#the axis=1 means that we are applying the function to each row, if we want to apply it to each column we would use axis=0
#worth mentioning that apply is slower than using vectorized operations like np.where or pd.cut
#native pandas functions are faster than apply, so try to use them when possible
```
